## Basic Program Structure
Understand the following parts
- import : load libary
- identifiers: names ( of variable, functions, etc. )
- statements
- builtins functions : helpful utilities that python provided, i.e. print
- keyword : construct statements, functions,  def.

```python
import math  # import statement is to load library math that others developed. Import is a keyword

r = int(input())  # identifier, builtins
area = math.pi*r*r 
print("area=", area) # builtin

def area(r):  # keyword def is used in define a function
    return math.pi*r*r

area(r)
```
